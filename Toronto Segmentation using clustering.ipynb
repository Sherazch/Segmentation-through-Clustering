 ### Let's Scrape the data using beautifulSoap

```python
webtext = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
```

```python
webtext
```

```python
bs = BeautifulSoup(webtext,'xml')
table = bs.find('div', class_='mw-parser-output')
```

```python
table
```

```python
table_rows = table.find_all('tr')
```

```python
table_rows
```

```python
cleaned_data=[]
for rows in table_rows:
    temp=[]
    for i in rows.find_all('td'):
        temp.append(i.text.strip())
    cleaned_data.append(temp)
cleaned_data
```

```python
type(cleaned_data)
```

### We have successfully scraped the data now let's convert it into a dataFrame

```python
#Initiating the data frame
df1=pd.DataFrame(data=cleaned_data)
```

```python
df2=df1[[0,1,2]]
```
